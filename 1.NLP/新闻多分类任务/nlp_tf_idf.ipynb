```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
train_df = pd.read_csv(r'E:\2.天池\NLP\新闻多分类任务\train_test_data\train_set.csv',sep='\t')
test_df = pd.read_csv(r'E:\2.天池\NLP\新闻多分类任务\train_test_data\test_a.csv',sep='\t')
```

```python
train_df
```

```python
test_df
```

```python
train_df['text'].iloc[:].values
```

```python
tfidf = TfidfVectorizer(max_features=3000).fit(train_df['text'].iloc[:].values)
```

```python
train_tfidf = tfidf.transform(train_df['text'].iloc[:].values)
test_tfidf = tfidf.transform(test_df['text'].iloc[:].values)
```

```python
train_tfidf
```

```python
from sklearn.linear_model import RidgeClassifier
clf = RidgeClassifier()
clf.fit(train_tfidf, train_df['gengxlabel'].iloc[:].values)
```

```python
df = pd.DataFrame()
df['label'] = clf.predict(test_tfidf)
df.to_csv('submit.csv', index=None)
```
